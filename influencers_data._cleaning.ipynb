```python
data=data[['name', 'headline', 'location', 'followers',
       'about', 'time_spent', 'content', 
       'media_type', 'num_hashtags',
       'hashtags', 'reactions', 'comments', 'votes']]
                       #Copy the necessary columns to reduce size
```

```python
data['headline'].unique()         #all the unique values of headline
```

```python
data.isna().sum()                 #checking missing values
data[data['location'].isna()]     #rows with value NA
```

```python
 #Filling blanks with NA and 0

data['location']=data['location'].fillna('NA')         #Filling blanks with NA
data['content']=data['content'].fillna('NA')
data['media_type']=data['media_type'].fillna('NA')
data['time_spent']=data['time_spent'].fillna('NA')
data['followers']=data['followers'].fillna(0)          #Filling blanks with 0

 
data['votes'] = data['votes'].str.replace(',', '')     #removing the , from the votes columns
data['votes']=data['votes'].fillna(0) 
```

```python
#changing the data type from float to int

data['followers'] = data['followers'].astype(int)  
data['votes']=data['votes'].astype(int)  
```

```python
#Cleaning the location column by removing "['] "symbols and splitting them up 

data['location'] = data['location'].str.replace('[', '').str.replace(']', '').str.replace(',', '').str.replace("'", "").str.split()
data['loc'] = data['location'].str[0]              #Selecting only the first word and creating a new column
data['loc']
#Use columns loc instead of location now
```

```python
#making a new column
data['reac_comm']=data['reactions']+data['comments']  
```

```python
#Sorting a column in descending

data=data.sort_values('reac_comm', ascending=False)
data
```

```python
#Pivot Table-> 'name', 'loc', 'followers'

pd.pivot_table(data,index=('loc','name'), values=('followers'),aggfunc='mean',margins=True)
```

```python
#Hashtags table

import re
hash=data[['name', 'num_hashtags', 'hashtags']]
hash['hashtags']=hash['hashtags'].str.replace('[', '').str.replace(']', '')
hash['hashtags'] = hash['hashtags'].apply(lambda x: re.findall(r'#\w+', x))
hash=hash[hash['num_hashtags'] != 0 ]
hash
 
```

```python
#Updating the dataset
data=data[['name', 'headline', 'loc', 'followers',
       'about', 'time_spent', 'content', 
       'media_type', 'num_hashtags',
       'hashtags', 'reactions', 'comments', 'votes', 'reac_comm']]


#downloading the data
data.to_csv('linkedinInflu.csv',index=False)
```
